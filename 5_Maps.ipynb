```python
import folium
import pandas as pd
from folium.plugins import MeasureControl
```

```python
def pintar_mapa(lat,lon,distance,number):
    
    # Ejecuto el mapa
    
    map1 = folium.Map(location=(lat,lon),tiles='Stamen Terrain',zoom_start=13)    

    # Airbnb
    
    marker_airbnb = folium.Marker(location=(lat,lon), popup="AirBnB", icon = folium.Icon(icon='home', color= "blue", prefix='fa')).add_to(map1)   

    # Puntos fijos
    
    marker_puerta_sol = folium.Marker(location=(40.4146500,-3.7004000), popup="Puerta del Sol", icon = folium.Icon(icon='flag', color= "red", prefix='fa')).add_to(map1)

    # Cargo el dataset
    
    data = pd.read_csv("KM.csv")
    
    # Museos de arte
    
    lst_coord = data.loc[((data["Identificador"] == "Museo de Arte") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='paint-brush ', color= "green", prefix='fa')).add_to(map1)
    
    
    # Esculturas al aire libre
    
    lst_coord = data.loc[((data["Identificador"] == "Escultura Exterior") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='caret-up', color= "purple", prefix='fa')).add_to(map1)
    
    # Campos de futbol
    
    lst_coord = data.loc[((data["Identificador"] == "Estadio de FÃºtbol") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='soccer-ball-o', color= "gray", prefix='fa')).add_to(map1)
    
    # Museo de Historia
    
    lst_coord = data.loc[((data["Identificador"] == "Museo de Historia") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='institution', color= "cadetblue", prefix='fa')).add_to(map1)
    
    # Palacio
    
    lst_coord = data.loc[((data["Identificador"] == "Palacio") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='fort-awesome', color= "beige", prefix='fa')).add_to(map1)
    
    # Metro
    
    lst_coord = data.loc[((data["Identificador"] == "Metro Station") & (data["Distancia_km"] < distance))].sort_values("Distancia_km").nsmallest(number, 'Distancia_km')
    lst_lat = [lst_coord["Latitud"][i] for i in lst_coord.index]
    lst_lon = [lst_coord["Longitud"][i] for i in lst_coord.index]
    lst_nom = [lst_coord["Nombre"][i] for i in lst_coord.index]    
    
    for i in range(len(lst_lat)):
        coord = [lst_lat[i],lst_lon[i]]
        marker = folium.Marker(location=(coord), popup=lst_nom[i], icon = folium.Icon(icon='subway', color= "orange", prefix='fa')).add_to(map1)
    

    
                  
    
    
    return map1
```
