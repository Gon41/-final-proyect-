```python
# Creo una nueva columna para categorizar los precios

# Copio la columna "price" porque se va a modificar

data["new_price"] = data["price"] 

# Creo los bin (intervalos) y los nombres de los intervalos que voy a necesitar

# bins = [0,31,60,100,1000] 

# names = ["1. Menos de 30€","2. Entre 30€ y 60€","3. Entre 60€ y 100€","4. Más de 100€"]

bins = [0,20,35,50,70,100,1000] 

names = ["1. Menos de 20€","2. Entre 21€ y 35€","3. Entre 36€ y 50€","4. Entre 51€ y 70€","5. Entre 71€ y 100€","6. Más de 100€"]

data["price_intervals"] = pd.cut(data["new_price"], bins, labels=names)

data = data.drop(columns=["new_price"])


data.head(3)
```

```python
# Reordeno las columnas

data =data.reindex(columns=['id','host_id','neighbourhood_group','neighbourhood','room_type','price_intervals','price','latitude','longitude','minimum_nights','number_of_reviews','availability_365','host_name',"name","last_review", "reviews_per_month", "calculated_host_listings_count"])

# Limpio el dataframe

data = data.drop(["name","last_review", "reviews_per_month", "calculated_host_listings_count"], axis="columns")

data.head(3)
```

## Función de búsqueda de pisos

```python
# Función para buscar un piso en airbnb conforme a los parámetros: DISTRITO, TIPO, INTERVALO DE PRECIO, MINIMO DE NOCHES

def busquedas_piso(distrito, kind, min_price, max_price, minimum_nights):
    
    # Filtro
    lst_pisos = data.loc[((data["neighbourhood_group"] == distrito) & (data["room_type"] == kind) & ((data["price"] > min_price) & (data["price"] < max_price)) & (data["minimum_nights"] == minimum_nights))]
    
    # Ordenar en función del precio (ascendente) y la cantidad de reseñas
    lst_pisos = lst_pisos.sort_values(['price', 'number_of_reviews'], ascending=[True, False])
    len_pisos = len(lst_pisos)
    
    if len(lst_pisos) > 0:
        return lst_pisos
    
    else:
        return "No se ha encontrado resultados con esos requisitos. Varie algún parámetro"

busquedas_piso("Centro","Private room",10,20,1)
```

```python
print(len(busquedas_piso("Chamberí","Entire home/apt",15,25,3)))
```

```python
busquedas_piso("Vicálvaro","Hotel room",20,50,1)
```
