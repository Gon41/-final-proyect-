## 4. Soccer Stadium

```python
csv4 = pd.read_csv("Estadios_futbol.csv")

csv4 = csv4.drop(columns=["Unnamed: 0"])

# Elimino las filas que no me interesan

csv4 = csv4.drop(range(1, 5),axis=0)
csv4 = csv4.drop([6,9],axis=0)

csv4.head(3)
```

## 5. Spanish Restaurant

```python
csv5 = pd.read_csv("Spanish_Restaurant.csv")

csv5 = csv5.drop(columns=["Unnamed: 0"])

csv5.head(3)
```

## 6. Italian Restaurant

```python
csv6 = pd.read_csv("Italian_Restaurant.csv")

csv6 = csv6.drop(columns=["Unnamed: 0"])

csv6.head(3)
```

## Concatenate y Final CSV

```python
# Acumulo todos los csv limpios en una variable, los concateno y reseteo el índice

frames = [csv1,csv2,csv3,csv4,csv5,csv6]

result = pd.concat(frames)

result = result.reset_index()

result = result.drop(columns=["index"])

result
```
